```python
## here is a problem in featuure name, here some word like action and actions
## are diff which is not a good thing to get rid of this thing we will use
## stemming ,it converts ['loved','loving','love'] to ['love','love','love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(text) :
    l = []
    for i in text.split() :
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())
```

```python
df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = 'english',max_features = 5000)
```

```python
vect = cv.fit_transform(df['tags']).toarray()
```

```python
vect
```

```python
cv.get_feature_names()
```

```python
## as here dimensions are high(5000) so euclidean distance is not reliable so
## we will calculate cosine distance i.e angle between vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vect)
```

```python
sim[1]
```

```python
## now if a music is given than show top 5 similar music
```

```python
def similarmusic(music) :
    ind = df[df['track_name'] == music].index[0]
    dis=sim[ind]
    recom_list = sorted(list(enumerate(dis)),reverse = True,key = lambda x : x[1])[1:6]
    for i in recom_list :
        print(df.iloc[i[0]].track_name)
    
```

```python
similarmusic('cry')
```
